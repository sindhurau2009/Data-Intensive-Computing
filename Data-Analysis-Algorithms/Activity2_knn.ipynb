```r
data1 <- na.omit(data)
```

```r
data$Sex <- as.numeric(data$Sex)
data$Housing <- as.numeric(data$Housing)
data$Saving.accounts <- as.numeric(data$Saving.accounts)
data$Checking.account <- as.numeric(data$Checking.account)
data$Purpose <- as.numeric(data$Purpose)
data$Credit.Risks <- factor(data$Credit.Risks)
```

```r
names(data)
data
```

```r
n.points <- 1000
```

```r
sampling.rate <- 0.8
```

```r
num.test.set.labels <- n.points * (1 - sampling.rate)
```

```r
training <- sample(1:n.points, sampling.rate * n.points, replace=FALSE)
training <- na.omit(training)
train <- subset(data[training, ],select =c(Age, Sex, Job, Housing, Credit.amount, Duration, Purpose))
```

```r
testing <- setdiff(1:n.points, training)
testing <- na.omit(testing)
test <- subset(data[testing, ], select = c(Age, Sex, Job, Housing, Credit.amount, Duration, Purpose))
```

```r
cl <- data$Credit.Risks[training]
factor(cl)
```

```r
true.labels <- data$Credit.Risks[testing]
true.labels
```

```r
install.packages("class",repos='http://cran.us.r-project.org')
library('class')
```

```r
knn(train, test, cl, k=3)
```

```r
df <- data.frame(k=numeric(20),train_80=numeric(20),train_70=numeric(20),train_75=numeric(20),train_60=numeric(20))
df
```

```r
for (k in 1:20) {
print(k)
predicted.labels <- knn(train, test, cl, k)
# We're using the R function knn()
num.incorrect.labels <- sum(predicted.labels != true.labels)
misclassification.rate <- num.incorrect.labels /
num.test.set.labels
    df$train_80[k] <- misclassification.rate*100
    print(misclassification.rate*100)
}
```

Misclassification rate for various k values are obtained from the loop above. This is for training set (80%). The smallest misclassification rate is 29.2, when k=12.
