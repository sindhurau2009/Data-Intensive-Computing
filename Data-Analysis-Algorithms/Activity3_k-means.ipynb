```r
data <- read.csv("File1.csv")
```

```r
data
```

```r
x2 <- data.frame(data$sex,data$lang,data$form,data$date3a,data$date3b, data$date4a)
x2 = na.omit(x2)
x2
```

```r
xa <- data.frame(x2$data.sex,x2$data.lang,x2$data.form,x2$data.date3a,x2$data.date3b)
xa
```

```r
ya <- x2$data.date4a
```

```r
kc1 <- kmeans(xa,3)
```

```r
kc1
```

```r
table(ya,kc1$cluster)
```

```r
plot(xa[c("x2.data.date3a","x2.data.date3b")], col=kc1$cluster)
points(kc1$centers[,c("x2.data.date3a", "x2.data.date3b")], col=1:3, pch=23, cex=3)
```

Considering the attributes sex, lang, form, date3a, date3b and predicting date4a value using the clusters generated, the above plot has been generated. However, it is not in the ideal format as no clusters are generated well.

```r
x1 <- data.frame(data$sex,data$cregion,data$state,data$game1,data$emplnw, data$game2d)
x1 = na.omit(x1)
x1
```

```r
y <- x1$data.game2d
```

```r
kc <- kmeans(x,3)
```

```r
kc
```

```r
table(y,kc$cluster)
```

```r
plot(x[c("x1.data.state","x1.data.emplnw")], col=kc$cluster)
points(kc$centers[,c("x1.data.state", "x1.data.emplnw")])
```

For k=3, 3 clusters have been generated using variables Sex, Cregion, State, Game1, Emplnw through which we can classify game2d values.
